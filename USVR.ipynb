```python
X = videos_matrix.values.T
X.shape
```

```python
SVD = TruncatedSVD(n_components=12, random_state=100)
matrix = SVD.fit_transform(X)
matrix.shape
```

```python
import warnings
warnings.filterwarnings("ignore",category =RuntimeWarning)
corr = np.corrcoef(matrix)
corr.shape
```
